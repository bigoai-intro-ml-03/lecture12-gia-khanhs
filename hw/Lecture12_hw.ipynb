# Tìm hiểu về Hyperparameter Tuning

**Hyperparameter tuning** là quá trình tối ưu hóa các tham số không học được trực tiếp từ dữ liệu trong quá trình huấn luyện mô hình, mà phải được cài đặt trước. 

Các phương pháp và kỹ thuật phổ biến trong hyperparameter tuning bao gồm:
+ **GridSeach**: Đây là phương pháp đơn giản nhưng mạnh mẽ, trong đó tất cả các kết hợp có thể của các giá trị hyperparameter trong một lưới (grid) được thử nghiệm, và kết quả tốt nhất được chọn. Tuy nhiên, Grid Search tốn nhiều thời gian và tài nguyên khi số lượng hyperparameter hoặc phạm vi giá trị lớn.
+ **RandomSearch**: Thay vì thử tất cả các kết hợp, Random Search chọn ngẫu nhiên một tập hợp các hyperparameter trong một khoảng giá trị cho trước. Phương pháp này thường nhanh hơn Grid Search và có thể tìm thấy một giải pháp gần tối ưu.
