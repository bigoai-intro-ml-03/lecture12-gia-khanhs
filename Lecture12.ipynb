# $Dataset$: [Red Wine Quality](https://www.kaggle.com/datasets/uciml/red-wine-quality-cortez-et-al-2009)

Bộ dữ liệu "Red Wine Quality" cung cấp các thông số phân tích của rượu vang từ vùng Vinho Verde ở Bồ Đào Nha (Để biết thêm chi tiết, ta có thể tham khảo ở  Cortez et al., 2009).

Bộ dữ liệu này cung cấp một cái nhìn toàn diện về các yếu tố hóa học có ảnh hưởng đến chất lượng của rượu vang, thông qua các đặc điểm sau:

* Input features:
1. fixed acidity (Độ axit tổng cố định)
2. volatile acidity (Độ axit bay hơi)
3. citric acid 
4. residual sugar (đường dư) 
5. chlorides (clo)
6. free sulfur dioxide (Lưu huỳnh dioxit tự do)
7. total sulfur dioxide (Lưu huỳnh dioxit tổng)
8. density (mật độ)
9. pH
10. sulphates 
11. alcohol (nồng độ cồn)

* Output: Quality score (between 0 -> 10)

# Ensemble

Ensemble Learning là một kỹ thuật trong Machine Learning mà trong đó nhiều mô hình (thường được gọi là "weak learners") được kết hợp lại với nhau để tạo ra một mô hình mạnh mẽ hơn, được gọi là "strong learner". Mục tiêu chính của Ensemble Learning là cải thiện độ chính xác của dự đoán và giúp mô hình có khả năng tổng quát hóa tốt trên dữ liệu mới.

Có ba phương pháp ensemble chính thường được sử dụng:
+ Bagging (Bootstrap Aggregating)
+ Boosting
+ Stacking

## Bagging method

**Bagging** dựa trên ý tưởng lấy mẫu có hoàn lại (bootstrap) từ tập dữ liệu huấn luyện để tạo ra nhiều tập dữ liệu con và sau đó huấn luyện cùng một loại model trên mỗi tập dữ liệu con này. Cuối cùng, kết quả từ tất cả các model này sẽ được tổng hợp lại (thông qua lấy trung bình hoặc bỏ phiếu) để tạo ra dự đoán cuối cùng.

````python
from sklearn.ensemble import BaggingClassifier
````

### BaggingClassifier
**Các tham số của BaggingClassifier**:
+ estimator: mô hình muốn ensemble
+ n_estimator: số lượng mô hình cần được tạo trong quá trình bagging
+ random_state: cố định trạng thái random


````python
bagg_model_dt = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=15, random_state=11)
bagg_model_dt.fit(X_train, y_train)
bagg_model_dt.score(X_test, y_test)
````

````python
bagg_model_svm = BaggingClassifier(estimator=SVC(), n_estimators=11, random_state=11)
bagg_model_svm.fit(X_train, y_train)
bagg_model_svm.score(X_test, y_test)
````

````python
bagg_model_svm = BaggingClassifier(estimator=LogisticRegression(max_iter=3000), n_estimators=15, random_state=11)
bagg_model_svm.fit(X_train, y_train)
bagg_model_svm.score(X_test, y_test)
````

## Boosting Method

**Boosting** là một nhóm các thuật toán mà ở đó các mô hình được huấn luyện tuần tự, với mỗi mô hình gắng sức để sửa lỗi của mô hình trước đó. Mỗi mô hình sau sẽ chú ý nhiều hơn tới những trường hợp mà mô hình trước đó đã dự đoán sai bằng cách tăng trọng số cho chúng. Cuối cùng, dự đoán từ tất cả các mô hình sẽ được tổng hợp lại để tạo ra dự đoán cuối cùng.

### Adaboost - Adaptive Boosting

**Các tham số của AdaBoostClassifier**:
+ estimator: mô hình muốn ensemble.
+ n_estimator: Số lượng mô hình học yếu (stumps) tối đa cần được tạo.
+ random_state: cố định trạng thái random.
+ learning_rate: Trọng số được áp dụng cho mỗi mô hình ở mỗi lần lặp, để kiểm soát tốc độ cập nhật trọng số của dữ liệu.

````python
from sklearn.ensemble import AdaBoostClassifier
````

````python
adaboost_svm = AdaBoostClassifier(DecisionTreeClassifier(), n_estimators=100, random_state=11)
adaboost_svm.fit(X_train, y_train)

adaboost_svm.score(X_test, y_test)
````

### GradientBoosting

GradientBoosting là 1 phiên bản cụ thể của Boosting sử dụng cây quyết định làm mô hình cơ bản.


**Các tham số của GradientBoostingClassifier**:
+ n_estimators: Số lượng "rounds" boost. Mỗi round thêm một mô hình vào chuỗi.
+ random_state: cố định trạng thái random.
+ learning_rate: Trọng số được áp dụng cho mỗi mô hình ở mỗi lần lặp, để kiểm soát tốc độ cập nhật trọng số của dữ liệu.
+ các tham số khác tương tự như RandomForest.

````python
# XGBoost: sử dụng gradient boosting.
# Mục tiêu: Tối ưu hóa cả hiệu suất model và tốc độ thực thi
from sklearn.ensemble import GradientBoostingClassifier

xgboost = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=11)
xgboost.fit(X_train, y_train)

xgboost.score(X_test, y_test)
````

## Stacking Methods

Trong **Stacking**, nhiều mô hình (cơ bản) khác nhau được huấn luyện độc lập với nhau từ cùng một tập dữ liệu huấn luyện. Sau khi hoàn thành quá trình huấn luyện, một mô hình mới (gọi là meta-learner hoặc mô hình cấp cao) sẽ được huấn luyện dựa trên các dự đoán (chứ không phải dữ liệu ban đầu) của các mô hình cơ bản. Mô hình meta-learner này sẽ học cách kết hợp tốt nhất các dự đoán từ các mô hình cơ bản để đưa ra dự đoán cuối cùng.

**Các tham số của StackingClassifier**:
+ estimators: Danh sách các mô hình sơ cấp (base learners). Thường được cung cấp dưới dạng một danh sách tuple, mỗi tuple chứa tên (string) và instance của mô hình.
+ final_estimator: Mô hình thứ cấp (meta-learner) dùng để kết hợp dự đoán từ các mô hình sơ cấp. Nếu không được chỉ định, mặc định thường là LogisticRegression trong sklearn.


````python
from sklearn.ensemble import StackingClassifier

estimators = [
  ('rf', DecisionTreeClassifier(min_samples_leaf= 5, max_features= None, max_depth= None, criterion= 'gini')),
  ('svr', SVC(random_state=11))]
final_estimator = LogisticRegression(max_iter=3000, C= 1, penalty= 'l2')
stacking = StackingClassifier(estimators, final_estimator)

stacking.fit(X_train, y_train)
stacking.score(X_test, y_test)
````

Dựa vào accuracy score trên test, ta thấy rằng việc sử dụng ensemble mang lại kết quả tốt hơn so với việc chỉ sử dụng đơn lẻ 1 model.

# Hyperparameter-tuning

**Nhận xét**:
Khi sử dụng các mô hình học máy mà giữ nguyên các cài đặt mặc định:
+ Kết quả thu được thường không thể đạt được cao nhất.
+ Có nguy cơ dẫn đến **Overfitting** và **Underfitting** (Decision Tree đạt 100% cho tập train và chỉ 60% cho tập test, đây là dấu hiệu của overfitting).
+ Ngoài ra có thể dẫn đến tiêu tốn nhiều tài nguyên tính toán mà hiệu suất không cao.

**Hyperparameter-tuning** là quá trình tối ưu hóa một mô hình machine learning bằng cách điều chỉnh các *siêu tham số* (hyperparameter) để cải thiện hiệu suất của nó trên tập dữ liệu cụ thể. 

Trong hyperparameter-tuning, ta sẽ tìm các bộ siêu tham số tối ưu như: 
+ Learning rate trong gradient descent
+ Số lượng epochs
+ Kích thước batch
+ Tham số liên quan đến cấu trúc của mô hình
+ ...

**Hyperparameter tuning trong sklearn**: sklearn cũng cung cấp những kỹ thuật thông dụng trong fine-tuning, phổ biến nhất là Grid Search và Randomized Search.


## GridSearch

GridSearch là một kỹ thuật phổ biến trong việc tìm kiếm không gian của các hyperparameters với mục tiêu tìm bộ hyperparameters tốt nhất cho một mô hình học máy nhất định. 

Đây là một phương pháp brute-force, tức là nó thử mọi tổ hợp có thể của hyperparameters để tìm ra bộ tối ưu nhất dựa trên một tiêu chí đánh giá cụ thể.

**Cách hoạt động**:  
+ Truyền vào một tập các hyperparameters và các giá trị tiềm năng cho mỗi hyperparameter cần thử nghiệm.
+ Thử nghiệm tất cả tổ hợp.
+ Chọn Bộ Hyperparameters tốt nhất.

**Ưu điểm**: 
+ GridSearch đảm bảo rằng từng tổ hợp của các hyperparameters đều được thử nghiệm, giúp tìm ra bộ cấu hình tối ưu nhất.

**Nhược điểm**:
+ Tốn Kém Về Mặt Tính Toán (do brute-force).

**Tham số của GridSeach**:
+ estimator: mô hình muốn fine tune.
+ param_grid: một từ điển (dictionary) hoặc một danh sách các từ điển, với mỗi *key* là tên của hyperparameter và *value* là danh sách các giá trị mà bạn muốn thử nghiệm cho hyperparameter đó.
+ scoring: tiêu chí đánh giá mô hình ('accuracy', 'f1', 'precision', 'recall', ...).
+ cv: cross-validation.
+ verbose: mức độ chi tiết mà GridSearchCV sẽ in ra trong quá trình thực hiện (giá trị cao hơn nghĩa là thông báo nhiều hơn).

````python
from sklearn.model_selection import GridSearchCV
````

### SVM

````python
from sklearn.svm import SVC

svc=SVC() 

parameters = [ 
              {'C':[1, 10, 100], 'kernel':['linear']},
               {'C':[1, 10, 100], 'kernel':['rbf'], 'gamma':[0.8, 1, 1.2]},
            #    {'C':[1], 'kernel':['poly'], 'gamma':["scale"], 'degree':[2, 3, 4,5,6,7]}
            ]

svc_cv = GridSearchCV(estimator = svc,  
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 5,
                           verbose=0)
svc_cv.fit(X_train, y_train)

print("Logistic Regression Parameters: {}".format(svc_cv.best_params_))
print("Best score is {}".format(svc_cv.best_score_))
````

### Logistic Regression

````python
from sklearn.linear_model import LogisticRegression
 
c_space = np.logspace(-10, 10, 30)
param_grid = {'C': c_space, 
              "penalty": ["l2"],
              "max_iter": [5000],
              }
 
LR = LogisticRegression()
 
LR_cv = GridSearchCV(LR, param_grid, cv = 5)
 
LR_cv.fit(X_train, y_train)
 
print("Logistic Regression Parameters: {}".format(LR_cv.best_params_))
print("Best score is {}".format(LR_cv.best_score_))
````

### Decision Tree

````python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()

param_tree = {"max_depth": np.arange(1, 10),
              "max_features": list(range(3, 10)),
              "min_samples_leaf": list(range(5, 50,5)),
              "min_samples_split": list(range(2, 10,2)),
            }

tree_cv = GridSearchCV(tree, param_tree, cv=5)

tree_cv.fit(X_train, y_train)

print("Decision Tree Parameters: {}".format(tree_cv.best_params_))
print("Best score is {}".format(tree_cv.best_score_))
````

````python
svc = SVC(**svc_cv.best_params_)
svc.fit(X_train, y_train)
print("Accuracy score của SVC cho tập test:",svc.score(X_test, y_test))
print("Accuracy score của SVC cho tập train:",svc.score(X_train, y_train))
LR = LogisticRegression(**LR_cv.best_params_)
LR.fit(X_train, y_train)
print("Accuracy score của Logistic Regression cho tập test:",LR.score(X_test, y_test))
print("Accuracy score của Logistic Regression cho tập train:",LR.score(X_train, y_train))
DT = DecisionTreeClassifier(**tree_cv.best_params_)
DT.fit(X_train, y_train)
print("Accuracy score của Decision Tree cho tập test:",DT.score(X_test, y_test))
print("Accuracy score của Decision Tree cho tập train:",DT.score(X_train, y_train))
````

Từ kết quả trên, ta có thể thấy rằng, kết quả sau khi sử dụng GridSearch tốt hơn so với model thông thường và cũng khắc phục overfitting.

## RandomizedSearchCV

RandomizedSearchCV là một kỹ thuật tối ưu hóa hyperparameters khác, giúp tìm kiếm không gian của các hyperparameters để tìm bộ hyperparameters tốt nhất cho một mô hình học máy. 

Khác với GridSearchCV, RandomizedSearchCV không thử hết tất cả các tổ hợp có thể mà thay vào đó chọn ngẫu nhiên một số tổ hợp để thử nghiệm.

**Cách hoạt động**:  
+ Truyền vào một tập các hyperparameters và các giá trị tiềm năng cho mỗi hyperparameter cần thử nghiệm.
+ Chọn ngẫu nhiên một số tổ hợp: Thay vì thử tất cả các tổ hợp, RandomizedSearchCV chọn ngẫu nhiên một số lượng tổ hợp hyperparameters xác định trước để thử nghiệm.
+ Chọn bộ Hyperparameters tốt nhất.

**Ưu điểm**: 
+ Hiệu Quả Hơn Về Mặt Tính Toán: Do chỉ thử nghiệm một số lượng tổ hợp ngẫu nhiên thay vì tất cả các tổ hợp có thể.

**Nhược điểm**:
+ Không Đảm Bảo Tìm Ra Bộ Tối Ưu Nhất: Vì chỉ thử nghiệm một số tổ hợp ngẫu nhiên nên có thể bỏ lỡ bộ cấu hình tối ưu nhất.

### SVM

````python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC

svc=SVC() 

parameters = [ {'C':[1, 10, 100], 'kernel':['linear']},
               {'C':[1, 10, 100], 'kernel':['rbf'], 'gamma':[0.8, 1, 1.2]},
            ]

grid_search = RandomizedSearchCV(svc, parameters, cv = 5)
grid_search.fit(X_train, y_train)

print("Tuned Logistic Regression Parameters: {}".format(grid_search.best_params_))
print("Best score is {}".format(grid_search.best_score_))
````

````python
c_space = np.logspace(-10, 10, 30)
param_grid = {'C': c_space, 
              "penalty": ["l2"],
              "max_iter": [5000],
              }
 
LR = LogisticRegression()
 
LR_cv = RandomizedSearchCV(LR, param_grid, cv = 5)
 
LR_cv.fit(X_train, y_train)
 
print("Logistic Regression Parameters: {}".format(LR_cv.best_params_))
print("Best score is {}".format(LR_cv.best_score_))
````

````python
### Decision Tree
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
 
param_dist = {"max_depth": np.arange(1, 15),
              "max_features": list(range(3, 15)),
              "min_samples_leaf": list(range(2, 50,1)),
              "min_samples_split": list(range(2, 50,1)),
            }

tree = DecisionTreeClassifier()
 
tree_cv = RandomizedSearchCV(tree, param_dist, cv = 5)
 
tree_cv.fit(X_train, y_train)
 
print("Decision Tree Parameters: {}".format(tree_cv.best_params_))
print("Best score is {}".format(tree_cv.best_score_))
````

Ta thấy rằng, tốc độ thực thi của RandomizeSearchCV khá nhanh so với GridSearchCV, tuy nhiên, kết quả thì đôi khi không tốt như GridSearchCV.
